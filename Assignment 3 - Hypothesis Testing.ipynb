```python
CustomerOrderForm=pd.read_csv("C:\\Users\\shred\\Desktop\\Data Science\\Assignments\\Assignment 3 - Hypothesis Testing\\Costomer+OrderForm.csv")
```

```python
p=CustomerOrderForm
```

```python
p
```

```python
d = {'Error Free': 1, 'Defective': 0}
P=pd.DataFrame(p["Phillippines"].map(d))
P.loc[:,"Indonesia"]=p["Indonesia"].map(d)
P.loc[:,"Malta"]=p["Malta"].map(d)
P.loc[:,"India"]=p["India"].map(d)
P.head()
```

```python
a=pd.DataFrame([[]])
for i in list(range(4)):
    for j in list(range(4)):
        c = [stats.ttest_ind(a=P.iloc[:,i],b=P.iloc[:,j],equal_var=False)]
        a.loc[i+1,j+1]=round(c[0][1],3)
a=a.iloc[1:,:]
a#p-Values for differences among different regions.
#1-Philippines,2-Indonesia,3-Malta,4-India
```
