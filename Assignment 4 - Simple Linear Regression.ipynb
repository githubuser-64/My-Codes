```python
import pandas as pd
df=Delivery_time=pd.read_csv("C:\\Users\\shred\\Desktop\\Data Science\\Assignments\\Assignment 4 - Simple Linear Regression\\delivery_time.csv")
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df["Sorting Time"],df['Delivery Time'])
plt.title('Delivery Time vs. Sorting Time')
plt.ylabel('Delivery Time')
plt.xlabel('Sorting Time')
plt.show()
```

```python
df.boxplot(column=['Delivery Time'])#No Outliers in the data.
```

```python
df.corr()
```

```python
import seaborn as sns
sns.regplot(x="Sorting Time", y="Delivery Time", data=df)
```

```python
import statsmodels.api as sm
y = df['Delivery Time']
x = df['Sorting Time']
model = sm.OLS(y, x).fit()
print(model.summary())
fig = plt.figure(figsize=(12,8))
fig = sm.graphics.plot_regress_exog(model, 'Sorting Time', fig=fig)
```

```python
df['Predicted']=model.fittedvalues
df['Error']=df['Delivery Time']-df['Predicted']
df.head()
```
