```python
import pandas as pd
import numpy as np
df=pd.read_csv("C:\\Users\\shred\\Desktop\\Data Science\\Assignments\\Assignment 10 - Recommendation System\\book.csv",encoding='latin-1')
df.pop('Unnamed: 0')
df.head()
```

```python
b=np.sort(df['User.ID'].unique())
```

```python
df.shape,len(df['User.ID'].unique()),len(df['Book.Title'].unique()),len(df['Book.Rating'].unique())
```

```python
df.set_index("User.ID", inplace = True)
df.head()
```

```python
df = pd.pivot_table(df,index='User.ID',
                             columns='Book.Title',
                             values='Book.Rating')
```

```python
df.head()
```

```python
df.fillna(0, inplace=True);df.head()
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances(df.values,metric='cosine')
user_sim_df = pd.DataFrame(user_sim)
user_sim_df.head()
```

```python
#Set the index and column names to user ids 
user_sim_df.index = b
user_sim_df.columns = b
```

```python
user_sim_df.iloc[0:10, 0:10]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:10, 0:10]
```

```python
a=pd.DataFrame([])
for i in user_sim_df.columns:
    for j in user_sim_df.columns:
        if j < i:
            if user_sim_df.loc[i,j]>0:
                a.loc[i,j]=user_sim_df.loc[i,j]
a
```

```python
b=pd.DataFrame([])
for i in a.columns:
    for j in a.columns:
        if j < i:
            if user_sim_df.loc[i,j]>0:
                b.loc[str(i)+' & '+str(j),"Similarities"]=user_sim_df.loc[i,j]
```

```python
b.head()
```

```python
b.sort_values(by='Similarities',ascending=False).head()
```

###### Similarities = 1 and duplicates have been removed as they have read just one book, though the same book.

```python
#Most Similar Users.
c=b.sort_values(by='Similarities',ascending=False)
d=c[c['Similarities']<1]
d.head()
```
